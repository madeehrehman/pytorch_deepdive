## Creating tensors

| Function | Use case |
|----------|----------|
| `torch.tensor([...])` | From Python list (copies data) |
| `torch.zeros` / `ones` / `full` | Fill with constant |
| `torch.arange` / `linspace` | Sequences |
| `torch.randn` | Random from N(0,1) |
| `torch.eye(n)` | Identity matrix |

````python
a = torch.tensor([1, 2, 3])           # 1D
b = torch.zeros(2, 3)                 # 2×3 filled with 0
c = torch.randn(2, 3)                   # random
d = torch.arange(0, 10, 2)            # [0, 2, 4, 6, 8]
e = torch.eye(3)                      # 3×3 identity

for name, t in zip("abcde", [a, b, c, d, e]):
    print(f"{name}: shape={tuple(t.shape)}, dtype={t.dtype}, device={t.device}")
````

## Shape, rank, numel

- **shape** — size per dimension `(rows, cols)`
- **ndim** (rank) — number of dimensions
- **numel** — total elements = product of shape

````python
t = torch.randn(2, 3, 4)  # 2 batches, 3×4 each

print(f"shape  : {tuple(t.shape)}")
print(f"ndim   : {t.ndim}")
print(f"numel  : {t.numel()}  (= 2×3×4 = {2*3*4})")
print(f"size(1): {t.size(1)}  # length of dim 1")
````

## Reshape (same data, new view of dimensions)

`.reshape()` and `.view()` change shape without copying (usually).  
**Rule:** total elements must stay the same.

```python
x.reshape(2, -1)   # -1 = infer this dim
```

````python
x = torch.arange(12)
print(f"original: {tuple(x.shape)} → {x.tolist()}")

row = x.reshape(3, 4)
col = x.reshape(4, 3)
flat = row.reshape(-1)  # back to 1D

print(f"3×4:\n{row}")
print(f"flatten: {flat.tolist()}")
````

### 🧪 Quiz — before running the next cell

`torch.arange(6).reshape(2, -1)` — what will the shape be?

````python
t = torch.arange(6).reshape(2, -1)
print(f"shape: {tuple(t.shape)}")
print(t)
````
